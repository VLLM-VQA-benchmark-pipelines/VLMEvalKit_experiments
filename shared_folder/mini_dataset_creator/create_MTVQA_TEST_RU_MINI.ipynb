```python
import pandas as pd
```

```python
# Открываем файл data.tsv и загружаем его в DataFrame
df = pd.read_csv("/root/LMUData/MTVQA_TEST_local.tsv", sep="\t", index_col=0)
```

```python
df.head(10)
```

```python
# оставим только RU подмножество
mini_ru_df = df[(df["category"] == "RU")][:10]
```

```python
mini_ru_df
```

```python
import os
import shutil

# Получаем путь к текущему рабочему каталогу
current_directory = os.getcwd()

# Путь к папке images
images_folder = os.path.join(current_directory, "images")

# Создаем папку images, если она не существует
if not os.path.exists(images_folder):
    os.makedirs(images_folder)

# Проходим по каждому пути в столбце "image_path"
for image_path in mini_ru_df["image_path"]:
    # Получаем имя файла
    filename = os.path.basename(image_path)
    
    # Путь для сохранения файла в папке images
    destination_path = os.path.join(images_folder, filename)
    
    # Копируем файл
    shutil.copy(image_path, destination_path)
    print(f"Copied {image_path} to {destination_path}")

print("All images have been copied to the 'images' folder.")
```

```python
mini_ru_df["image_path"] = mini_ru_df["image_path"].str.replace(
    "/root/LMUData/images/MTVQA_TEST/", "images/"
)
```

```python
mini_ru_df
```

```python
mini_dataset_name = "MTVQA_TEST_RU_MINI"
mini_dataset_annot_file = f"{mini_dataset_name}.tsv"
mini_ru_df.to_csv(mini_dataset_annot_file, sep="\t", index=False)
```

```python
os.makedirs(mini_dataset_name, exist_ok=True)
shutil.move("images", mini_dataset_name)
```
